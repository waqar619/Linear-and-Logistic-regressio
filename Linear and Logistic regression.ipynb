```python
from sklearn.datasets import load_boston
```

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=load_boston()
```

```python
df
```

```python
pd.DataFrame(df.data)
```

```python
dataset=pd.DataFrame(df.data)
```

```python
dataset.columns=df.feature_names
dataset.head()
```

```python
dataset['price']=df.target
```

```python
dataset.head()
```

```python
## dividing the datasets into dependable and independable feature 
X=dataset.iloc[:,:-1] ## independent feature
y=dataset.iloc[:,-1] ## depenedent feature 
```

```python
X.head()
```

```python
y.head()
```

```python
!pip install matplotlib
```

```python
!pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

```python
## Linear regression 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg=LinearRegression()
mse=cross_val_score(lin_reg,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
mean_mse=np.mean(mse)
print(mean_mse)


```

```python
X.shape
```

```python
y.shape
```

```python
## ridge regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}
ridge_regressor=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=10)
ridge_regressor.fit(X_train,y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
## lasso regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}
lasso_regressor=GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=10)
lasso_regressor.fit(X_train,y_train)
```

```python
print(lasso_regressor.best_params_)

print(lasso_regressor.best_score_)
```

```python
y_pred=lasso_regressor.predict(X_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score1=r2_score(y_test,y_pred)
print(r2_score1)
```

# logistic regression


```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
df=load_breast_cancer()
df
```

```python
# independant feature
X=pd.DataFrame(df['data'],columns=df['feature_names'])
X
```

```python
# Dependent feature
y=pd.DataFrame(df['target'],columns=['Target'])
y
```

```python
y['Target'].value_counts()
```

```python
# Test train split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

```python
params=[{'C':[1,5,10]},{'max_iter':[100,150]}]
```

```python
model1=LogisticRegression(C=100,max_iter=100)
```

```python
model=GridSearchCV(model1,param_grid=params,scoring='f1',cv=5)
```

```python
model.fit(X_train,y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import  confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
